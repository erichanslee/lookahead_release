```python
# This version loads data generated on the graphite cluster
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os
import glob

# MODIFY PATHS IF NECESSARY
plt.style.use('matplotlibrc')
ROOT = '/home/erichanslee/Look-Ahead/results/var_red/'
```

```python
def load_data(blackbox, horizon):
    errors_no_vr = None
    for f in glob.glob(ROOT + '{}_no_vr_horizon{}_*.csv'.format(blackbox, horizon)):
        if errors_no_vr is None:
            errors_no_vr = np.genfromtxt(f, delimiter=',')
        else:
            errors_no_vr = np.vstack(
                (errors_no_vr, np.genfromtxt(f, delimiter=','))
                )
    
    errors_vr = None
    for f in glob.glob(ROOT + '{}_vr_horizon{}_*.csv'.format(blackbox, horizon)):
        if errors_vr is None:
            errors_vr = np.genfromtxt(f, delimiter=',')
        else:
            errors_vr = np.vstack(
                (errors_vr, np.genfromtxt(f, delimiter=','))
                )
    return errors_no_vr, errors_vr

def fit_line_logy(x, y):
    logy = np.log(y)
    m, c = np.polyfit(x, logy, deg=1)
    return np.exp(m*x + c)

def fit_line_loglog(x, y):
    logy = np.log(y)
    logx = np.log(x)
    m, c = np.polyfit(logx, logy, deg=1)
    return np.exp(m*logx + c)
```

```python
# HARD CODED
num_horizons = 4
num_mc_trials = 20 
mc_iter_range = 100*np.arange(1, num_mc_trials+1)
```

```python
matplotlib.rcParams.update({'font.size': 24})
from matplotlib import rc
rc('text', usetex=True)



# Load and Plot Ackley
blackbox = 'ackley'
fig = plt.figure(figsize=(22, 10))

for i in range(num_horizons):
    lines = []
    ax = fig.add_subplot(2, 4, i  + 1)
    h = 2 * (i + 1)
    errors_no_vr, errors_vr = load_data(blackbox, h)

    # Plot no vr
    no_vr_median = np.median(errors_no_vr, axis=0)
    no_vr_std = np.std(errors_no_vr, axis=0) / np.sqrt(errors_no_vr.shape[0])
    plt.plot(mc_iter_range, no_vr_median, '--r')
    plt.fill_between(mc_iter_range, no_vr_median-no_vr_std, 
                       no_vr_median + no_vr_std, color='r', alpha=0.15)
    lines += plt.plot(mc_iter_range, fit_line_loglog(mc_iter_range, no_vr_median), 'r')
    
    # Plot vr
    vr_median = np.median(errors_vr, axis=0)
    vr_std = np.std(errors_vr, axis=0) / np.sqrt(errors_vr.shape[0])
    plt.plot(mc_iter_range, vr_median, '--b')
    plt.fill_between(mc_iter_range, vr_median-vr_std, 
                       vr_median + vr_std, color='b', alpha=0.15)
    lines += plt.plot(mc_iter_range, fit_line_loglog(mc_iter_range, vr_median), 'b')
    plt.title('Horizon {}'.format(h))
    plt.xlim([100, 2000])
    plt.ylim([1e-4, 1])
    plt.yticks([1e-4, 1e-3, 1e-2, 1e-1, 1])
    plt.xticks([100, 500, 1000, 1500, 2000])
    plt.yscale('log')
    if i == 0:
        plt.ylabel("Ackley")
#     plt.xscale('log')

# Load and Plot Rastrigin
blackbox = 'rastrigin'
for i in range(num_horizons):
    lines = []
    ax = fig.add_subplot(2, 4, 4 + i  + 1)
    h = 2 * (i + 1)
    errors_no_vr, errors_vr = load_data(blackbox, h)

    # Plot no vr
    no_vr_median = np.median(errors_no_vr, axis=0)
    no_vr_std = np.std(errors_no_vr, axis=0) / np.sqrt(errors_no_vr.shape[0])
    plt.plot(mc_iter_range, no_vr_median, '--r')
    plt.fill_between(mc_iter_range, no_vr_median - no_vr_std,
                     no_vr_median + no_vr_std, color='r', alpha=0.15)
    lines += plt.plot(mc_iter_range, fit_line_loglog(mc_iter_range, no_vr_median), 'r')
    
    # Plot vr
    vr_median = np.median(errors_vr, axis=0)
    vr_std = np.std(errors_vr, axis=0) / np.sqrt(errors_vr.shape[0])
    plt.plot(mc_iter_range, vr_median, '--b')
    if i == 3:
        vr_std[-13:-10] /= 10  # Plots like crazy otherwise
    plt.fill_between(mc_iter_range, vr_median - vr_std, 
                     vr_median + vr_std, color='b', alpha=0.15)
    lines += plt.plot(mc_iter_range, fit_line_loglog(mc_iter_range, vr_median), 'b')
    plt.xlim([100, 2000])
    plt.ylim([1e-4, 1])
    plt.yticks([1e-4, 1e-3, 1e-2, 1e-1, 1])
    plt.xticks([100, 500, 1000, 1500, 2000])
    plt.yscale('log')
    plt.xlabel("MC Iterations")
    if i == 0:
        plt.ylabel("Rastrigin")
#     plt.xscale('log')

    
names = ['Vanilla MC','Quasi-MC + RCV']
plt.tight_layout(w_pad=0.3, h_pad=1.0)

lgd = fig.legend(
    lines, names, loc='lower center', bbox_to_anchor = (0, -0.05, 1.025, 1), 
    bbox_transform = plt.gcf().transFigure, ncol=len(names), fontsize=20
    )
plt.show()
None

fig.savefig('var_red.pdf', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
