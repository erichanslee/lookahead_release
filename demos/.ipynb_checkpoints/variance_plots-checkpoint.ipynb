```python
def plot_gp(ax, gp, xtrain, ytrain):
    X = np.linspace(0,1,num=100)
    X = X[:,np.newaxis]
    sigma = np.sqrt(gp.variance(X))
    Y = gp.mean(X)
    ax.fill(np.concatenate([X, X[::-1]]),
         np.concatenate([Y - 1.9600 * sigma,
                        (Y + 1.9600 * sigma)[::-1]]),
         alpha=.5, fc='m', ec='None', label='95% confidence interval')
    ax.plot(X,Y,'-r')
    ax.plot(xtrain ,ytrain,'ko', linewidth=3,  markersize=6)

def plot_acquisition(ax, acq):
    X = np.linspace(0,1,num=100)
    X = X[:,np.newaxis]
    acq_values = acq.evaluate_at_point_list(X)
    acq_values = acq_values - np.min(acq_values)
    acq_values = acq_values / np.max(acq_values)
    ax.plot(X,acq_values, '--g', linewidth=2.5)
    ax.plot(X[np.argmax(acq_values), :], np.max(acq_values), 'b*', markersize=4)
    ax.grid()
    
def normalize_acq(array_in):
    return array_in / np.max(array_in)

def plot_stuff(ax, gp, ei, title=None):
    X = np.linspace(0, 1, 100)[:, None]
    mu = gp.mean(X)
    sigma = gp.variance(X)
    ei_vals = ei.evaluate_at_point_list(X)
    ms = 5
    mss = 13
    l1 = ax.plot(X, f(X), 'k', lw=3)
    l2 = ax.plot(X, mu, '-r', lw=3)
    l3 = ax.fill(np.concatenate([X, X[::-1]]),
             np.concatenate([mu - 1.9600 * sigma,
                            (mu + 1.9600 * sigma)[::-1]]),
             alpha=.3, fc='m', ec='None', label='95% confidence interval')
    l4 = ax.plot(x_train, y_train, 'ok', markersize=ms)
    ax2 = ax.twinx()
    l5 = ax2.plot(X, ei_vals, 'g:', lw=2.5)
    l6 = ax2.plot(X[np.argmax(ei_vals), :], ei_vals[np.argmax(ei_vals)], 'g*', markersize=mss)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax2.get_xaxis().set_visible(False)
    ax2.get_yaxis().set_visible(False)
    if title is not None:
        ax2.set_title(title)
    lines = l1 + l2 + l3 + l4 + l5 + l6
    names = ['Objective', 'GP mean', 'GP uncertainty', 'Observations', 'Acquisition', 'Argmax']
    plt.legend(
        lines, names, loc='lower center', bbox_to_anchor = (0, 0.02, 1.025, 1), 
        bbox_transform = plt.gcf().transFigure, ncol=len(names), fontsize=8.5
        )
```

```python

num_trials = 2
xtrain = np.array([0.4/3, 2.6/3])[:, None]
ytrain = np.array([0, 0])
gp = GaussianProcessSimple(xtrain, ytrain)
gp.set_hypers([2,0.1])
domain = TensorProductDomain([ClosedInterval(0, 1)])
```

```python
# get an error estimate for vanilla BO, using horizon
def get_errors(gp, domain, horizon, num_repetitions, num_mc_trials):
    eval_point = np.array([[0.5]])
    error_no_vr = np.zeros(num_mc_trials)
    error_vr = np.zeros(num_mc_trials)

    # Assume the true value is estimated with a lot of mc iters
    rollout_ei_vr = RolloutEI_VR_AB(gp, domain, h, mc_iters=int(3e3))
    true_value = rollout_ei_vr.evaluate_at_point_list(eval_point)
    
    for _ in range(num_repetitions):
        value_no_vr = np.zeros(num_mc_trials)
        value_with_vr = np.zeros(num_mc_trials)
        mc_iter_range = 100*np.arange(1, num_mc_trials+1)

        # Loop
        for i in range(num_mc_trials):
            mc_iters = mc_iter_range[i]
            rollout_ei = RolloutEI(gp, domain, h, mc_iters=mc_iters, opt_mode='grid')
            value_no_vr[i] = rollout_ei.evaluate_at_point_list(eval_point)
        error_no_vr += np.abs(value_no_vr-true_value)

    # Loop (don't need multiple repetitions, mc is determinstic)
    value_with_vr = np.zeros(num_mc_trials)
    for i in range(num_mc_trials):
        mc_iters = mc_iter_range[i]

        # Sample full variance reduction
        rollout_ei_vr = RolloutEI_VR_AB(gp, domain, h, mc_iters=mc_iters)
        value_with_vr[i] = rollout_ei_vr.evaluate_at_point_list(eval_point)
    error_vr  = np.abs(value_with_vr-true_value)
    return mc_iter_range, error_no_vr / num_repetitions, error_vr / num_repetitions
```

```python
num_repetitions = 10
num_mc_trials = 10
num_horizons = 2

fig, ax = plt.subplots(1, num_horizons, figsize=(3*num_horizons, 3))
errors_no_vr = np.zeros(())

for h in range(2, num_horizons + 2):
    mc_iter_range, error_no_vr, error_vr = get_errors(gp, domain, h, num_repetitions, num_mc_trials)
    ax[h-2].plot(mc_iter_range, error_no_vr)
    ax[h-2].plot(mc_iter_range, error_vr)

```
