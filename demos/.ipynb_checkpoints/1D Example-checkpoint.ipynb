```python
'''
This notebook will test a number of a few different things needed required for the lookahead acquisition function.
In particular, it will first run through the Gaussian process wrapper used. 

The lookahead acquisition function we consider has the form:
\int_{-\infty}^{\infty} loss(z_2) p(F_q = z_q) dz_q

F_q is the lookahead random variable representing possible function values after q-1 steps of Bayesian optimization.
In the case of q=1, p(F_1 = z_1) has the pdf of the underlying GP's posterior, and we know what it is. The tricky
part is defining p(F_q = z_q) for q > 1. The purpose of look_ahead.py is to compute F_q for q=2, based off 
the density formulation given in: << Gaussian Processes for Global Optimization, Osborne et al. >>

The code is not yet optimized and is very much in an initial state. It's not worth polishing too much for now, 
as it will only be used to generate a few initial plots
'''

# Proper import statements. NOTE: this notebook uses gaussian_process_moe, which is a sigopt internal version.
# Any alternative GP used needs to match the interface in gaussian_process_interface.py
import numpy as np
from lookahead.utils import unit_grid_points2
from lookahead.gaussian_process_moe import GaussianProcessMOE as GaussianProcess #import different GP class if needed
from lookahead.expected_improvement import ExpectedImprovement
```

```python
# Calculate initial data
np.random.seed(1111)
ntrain = 2
xtrain = np.random.rand(ntrain,1)
xtrain[0,0] = 1/3
xtrain[1,0] = 2.1/3
f = lambda x : (x[:,0] - 0.5)**2
ytrain = f(xtrain)
```

```python
'''
Plot underlying function vs GP mean
'''

# Plot prediction vs actual function
import matplotlib.pyplot as plt

# Build a fictional GP of some function over a discrete, 10 x 10 grid
gp = GaussianProcess(xtrain,ytrain,kernel='Matern')
gp.train()

ntest = 30
X = np.linspace(0,1,num=ntest)
X = X[:,np.newaxis]
Y = f(X)
plt.subplot(1,2,1)
plt.plot(X,Y)
plt.title('True Function')

Y = gp.mean(X)
plt.subplot(1,2,2)
plt.plot(X,Y)
plt.plot(xtrain,ytrain,'ro')
sigma = np.sqrt(gp.variance_squared(X))
plt.plot(X, Y + 2*sigma)
plt.plot(X, Y - 2*sigma)
plt.title('GP Mean')
plt.show()

print(gp.covariance.get_hyperparameters())
```

```python
'''
Plot regular expected improvement contour
'''

ei = ExpectedImprovement(gp)
EI = ei.compute_acquisition_function(X)
plt.plot(X,EI)
plt.title('Expected Improvement')
plt.show()
```

```python
'''
Plot the pdf p(F_2 = z_2) at a single point xi. The goal here is to understand how the shape of p(F_q = z_q)
gets "warped" from a normal Gaussian. The distribution may be flat if the normal EI at that point is zero,
so be careful. Generally, the distribution may look gaussian, multimodal, or non-smooth. xi is chosen here
to maximize the expected improvement
'''

from look_ahead import look_ahead_two as LH2 
from utils import normpdf

# Plot probability distributions for one and two step look-ahead, at the point
# where the regular EI is highest

idx = np.argmax(EI)

xi = X[[idx],:]
mu = gp.mean(xi)
sigma = np.sqrt(gp.variance_squared(xi))
lb = mu-4*sigma; ub = mu+4*sigma
Z = np.linspace(lb,ub,num=50)
pdf_Z = normpdf(Z,mu,sigma)
idx = np.argmax(pdf_Z)
lh = LH2(gp)

plt.subplot(1,2,1)
plt.plot(Z,pdf_Z)
plt.plot(Z[idx],pdf_Z[idx],'ro')
plt.xlim((lb,ub))
plt.title('PDF for q=1 (Gaussian)')

plt.subplot(1,2,2)
lh_values = lh.look_ahead_prob(Z,xi)
plt.plot(Z,lh_values)
plt.plot(Z[idx],pdf_Z[idx],'ro')
plt.xlim((lb,ub))
plt.title('PDF for q=2 (Not Gaussian)')

plt.show()
```

```python
'''
Plot the pdf p(F_2 = z_2) at a single point xi. The goal here is to understand how the shape of p(F_q = z_q)
gets "warped" from a normal Gaussian. The distribution may be flat if the normal EI at that point is zero,
so be careful. Generally, the distribution may look gaussian, multimodal, unsymmetric
'''

from look_ahead import look_ahead_two as LH2 
from utils import normpdf

# Plot probability distributions for one and two step look-ahead, at the point of the median EI
idx = np.argsort(EI)[len(EI)//2]

xi = X[[idx],:]
mu = gp.mean(xi)
sigma = np.sqrt(gp.variance_squared(xi))
lb = mu-4*sigma; ub = mu+4*sigma
Z = np.linspace(lb,ub,num=50)
pdf_Z = normpdf(Z,mu,sigma)
idx = np.argmax(pdf_Z)
lh = LH2(gp,num_quad_points_inner=60)

plt.subplot(1,2,1)
plt.plot(Z,pdf_Z)
plt.plot(Z[idx],pdf_Z[idx],'ro')
plt.xlim((lb,ub))
plt.title('PDF for q=1 (Gaussian)')

plt.subplot(1,2,2)
lh_values = lh.look_ahead_prob(Z,xi)
plt.plot(Z,lh_values)
plt.plot(Z[idx],pdf_Z[idx],'ro')
plt.xlim((lb,ub))
plt.title('PDF for q=2 (Not Gaussian)')

plt.show()
```

```python
'''
Plot the contours for regular Expected Improvement and the q=2 Expected Improvement
'''

# Calculate look-ahead acquisition function *** WARNING: will take a few hours ***
LH = lh.compute_acquisition_function(X)

plt.subplot(1,2,1)
plt.plot(X,LH)
plt.title('Two-step Expected Improvement')

plt.subplot(1,2,2)
plt.plot(X,EI)
plt.title('Expected Improvement')

plt.show()
```
