```python
import matplotlib
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('matplotlibrc')
from experiments_data import load_data, plot_runs, plot_aggregate_runs
```

```python
blackboxes = ['branin','sixhump','ackley','rastrigin']
titles = ['Branin','Sixhump','Ackley','Rastrigin']
nb = len(blackboxes)
(fig, ax) = plt.subplots(1, nb, figsize=(12,3))
optimizers = ['portfolio_ei2','portfolio_ei4','ei', 'ucb0', 'ucb2', 'kg']

for i in range(nb):
    blackbox = blackboxes[i]
    lines = plot_runs(ax[i], optimizers, blackbox, offset=7, offset_end=25)
    ax[i].set_title(titles[i])
ax[0].set_ylabel('Min Value Observed')
names = ['PS2', 'PS4', 'EI', 'UCB-0', 'UCB-2', 'KG']
lgd = fig.legend(
    lines, names, loc='lower center', bbox_to_anchor = (0, -0.08, 1.025, 1), 
    bbox_transform = plt.gcf().transFigure, ncol=len(names), fontsize=10
    )
plt.tight_layout()
fig.savefig('policy_search.pdf', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
